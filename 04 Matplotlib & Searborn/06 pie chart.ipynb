```python
import matplotlib.pyplot as plt
plt.style.use("ggplot")  # Choose any style you like
```

```python
labels = ["Sachin","Sehwag","Kohli","Yuvraj"]
runs = [18000,8000,12000,9500]

plt.title("Carrer Runs of cricketers")
plt.pie(runs,labels=labels)
```

```python
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99']
plt.pie(runs, labels=labels, colors=colors)
```

```python
plt.pie(
    runs,
    labels=labels,
    colors=colors,
    wedgeprops={'edgecolor': 'black', 'linewidth': 2, 'linestyle': '--'}
)
plt.show()
```

```python
explode = [0.1,0,0,0]
plt.pie(runs,labels=labels, explode=explode, colors=colors, shadow = True, startangle=10,autopct='%1.1f%%')
plt.title("Explode pie example")
```
