```python
import matplotlib.pyplot as plt

days = [1,2,3,4,5,6]
study = [8,2,10,8,8,9]
gym = [2,1,1.5,3,1,2]
sleep = [8,10,9,12,10,10]
travel = [3,5,8,6,4,10]

labels = ['study', 'gym','sleep','travel']
colors = ['red','blue','yellow','green']

plt.stackplot(days,study,gym,sleep,travel, labels = labels, colors = colors, alpha= 0.4)
plt.title("Activities Tracker")
plt.xlabel("Days")
plt.ylabel("Hours")
plt.legend(loc = "upper left")
```
